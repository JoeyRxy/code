```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(-10, 10, 200)


def g(t):
    a = 1 - 4 * t ** 2
    b = np.cos(np.pi * t)
    return 4 * b / (np.pi * a)

y=g(t)
plt.plot(t, y)
```
